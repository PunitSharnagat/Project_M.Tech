### 2. For Drive

```python
data1 = []
for p, coords in cbd.cordinates.iteritems():
    
    location = tuple(map(float, coords[1:-1].split(', ')))
    dist = 1000     #cbd['distance'][i]
    city = cbd['city'][p]
    place = cbd['CBD'][p]
    network_type = 'drive'
    
    # Importing data from place using centre point as drivable network
    G = ox.graph_from_point(location, dist = dist, network_type = network_type)
    
    # convert MultiDiGraph to an undirected MultiGraph
    M = ox.get_undirected(G)
    
    # saving graph as nodes
    nodes = ox.graph_to_gdfs(G,edges=False)
    
    #Converting geodataframe (geopanda) to panda
    pd_nodes = pd.DataFrame(nodes)
    pd_nodes.to_csv(f"C:/Users/HP/Desktop/OSMnx Work/Mannual Circuity/Nodes/D_{city}_{place}_nodes.csv")
    
    #reading files from local directory
    nodes=pd.read_csv(f"C:/Users/HP/Desktop/OSMnx Work/Mannual Circuity/Nodes/D_{city}_{place}_nodes.csv")
    # calculation of circuity for Drive 
    data = []
    i=0
    while i < len(nodes):
        j = i
        # Making pairs of each stop with other stops
        while j < len(nodes):
            if j != i:
                origin = nodes['osmid'][i]
                dest = nodes['osmid'][j]

                # Origin and Destination coordinates
                O_coords = (nodes['y'][i], nodes['x'][i])    #(lat,Lon) format
                D_coords = (nodes['y'][j], nodes['x'][j])

                # calculating Network distance
                o = nodes['osmid'][i] 
                d = nodes['osmid'][j]
                od = ox.shortest_path(G,o,d, weight = 'length')
                if type(od) != list:
                    j=j+1
                    continue
                length = int(sum(ox.utils_graph.get_route_edge_attributes(G,od,'length')))

                # Calculating Euclidean distance
                distance = (geopy.distance.geodesic(O_coords, D_coords).m) #(lat,Lon) format

                # Circuity calculation
                circ = length/distance

                # saving data to dataframe
                data.append({'Origin':origin, 'Destination':dest,'O_coords': O_coords,'D_coords': D_coords,
                            'Network_dist':length, 'Euclidean_dist':distance, 'Circuity':circ})
                j=j+1
            else:
                j=j+1

        i = i+1
    Pairs = pd.DataFrame(data)
    Pairs = Pairs[Pairs['Circuity'] <5] #removing pairs with circuity  > 5
    # Calculating average circuity for network
    c=0
    temp=0
    for c in Pairs.index:
        temp=temp+Pairs['Circuity'][c]
    Avg_circuity = temp/len(Pairs)
    Avg_circuity
    
    # Saving to CSV file
    Pairs.to_csv(f"C:/Users/HP/Desktop/OSMnx Work/Mannual Circuity/Pairs/D_{city}_{place}_pairs.csv")
    
    # Circuity for all in one place
    data1.append({'City':city,'CBD':place,'Avg_Circuity':Avg_circuity})
    print(f"Done for {city},{place}; Avg circuity = {Avg_circuity}.")
    

Circuity = pd.DataFrame(data1)
Circuity.to_csv(f"C:/Users/HP/Desktop/OSMnx Work/Mannual Circuity/Circuity_{network_type}.csv")
```
